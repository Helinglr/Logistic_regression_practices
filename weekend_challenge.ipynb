```python
import numpy as np 
import pandas as pd

```

```python
test_data = np.array([[1, 5],    
    [2, 4],      
    [3, 3],    
    [4, 2],    
    [5, 1],    
    [2, 6],    
    [3, 4],    
    [4, 4],    
    [6, 3]])
```

```python
set = pd.read_csv('points.csv')
affects = pd.read_csv('affects.csv')
list_of_points = np.array(set)
list_of_points 
arr = list(list_of_points)
eliminated = []
list_of_points
```

```python
while len(arr) >0 : 


    ## referans noktası seçilme kısmı
    reference_point = np.array(arr.pop())
    
    ## difleri alma kısmı
    difference = list_of_points - reference_point
    
    ## feature sayısına göre eliminasyon kısmı

    for i in range(len(difference)):
        x= difference[i]
        if np.all(x*affects < 0):
            eliminated.append(list_of_points[i])
        elif (np.any(x * affects < 0) and  np.all(x * affects <= 0)):        # Hiçbir boyutta daha kötü değil
            eliminated.append(list_of_points[i])
    arr = list([point for point in arr if not any(np.all(point == p ) for p in eliminated)])## elenen noktaları seçim listesinden çıkarttık 
    list_of_points = np.array([point for point in list_of_points if not any(np.all(point == p) for p in eliminated)])
    
```
