```python
import numpy as np
import pandas as pd

# sigmoid fonksiyonu
def sigmoid(w, b, x):
    z = np.dot(x, w) + b
    sgm = 1 / (1 + np.exp(-z))
    return sgm

# cost fonksiyonu (log loss)
def current_cost(x, y, w, b):
    y_pred = sigmoid(w, b, x)
    # log10 yerine log kullanıyoruz
    loss = -y * np.log(y_pred + 1e-8) - (1 - y) * np.log(1 - y_pred + 1e-8)
    cost = np.mean(loss)
    return cost

# w ve b güncelleme
def update_w_b(x, y, m, n, alpha, w_i, b_i):
    y_pred = sigmoid(w_i, b_i, x)
    error = y_pred - y  # (m,)
    dw = np.dot(x.T, error) / m  # shape (n,)
    db = np.sum(error) / m
    w_i = w_i - alpha * dw
    b_i = b_i - alpha * db
    return w_i, b_i

# ana logistic regression modeli
def logistic_regression_model(x, y, m, n, w_i, b_i, alpha, epochs):
    for i in range(epochs):
        w_i, b_i = update_w_b(x, y, m, n, alpha, w_i, b_i)
        if i % 1000 == 0:
            print(f"iteration: {i}, cost value : {current_cost(x, y, w_i, b_i):.4f}")
    return w_i, b_i

# test fonksiyonu
def test(x_test, w, b):
    y_pred = sigmoid(w, b, x_test)
    return (y_pred >= 0.5).astype(int)  # sınıf tahmini

# tahminleri gerçeklerle karşılaştırma
def compute_values(y_pred, y_real):
    result = np.c_[y_pred, y_real]
    print("Predictions | Reals")
    print(result)


# --- Veri yükleme ---
data = pd.read_csv("framingham.csv")

y_train = data['TenYearCHD'].values
x_train = data[['male', 'age', 'education', 'currentSmoker', 'cigsPerDay', 'BPMeds',
       'prevalentStroke', 'prevalentHyp', 'diabetes', 'totChol', 'sysBP',
       'diaBP', 'BMI', 'heartRate', 'glucose']]

# NaN değerleri ortalama ile doldur
x_train = x_train.fillna(x_train.mean())

# Normalizasyon
mean = x_train.mean()
std = x_train.std()
x_train_norm = (x_train - mean) / std

# test seti
x_test = x_train.iloc[17:22, :]
x_test_norm = (x_test - mean) / std
y_test = y_train[17:22]

m, n = x_train.shape
print(m, n)

# model eğitimi
alpha = 0.01
epochs = 10000
w_i = np.zeros(n)
b_i = 0.0

w_r, b_r = logistic_regression_model(x_train_norm.values, y_train, m, n, w_i, b_i, alpha, epochs)

# test
predictions = test(x_test_norm.values, w_r, b_r)
compute_values(predictions, y_test)
```
