```python
## sigmoid function 
# g(z) = 1/ (1+e^(-z)) and z = w*x+b w and x are vectors 
## g(z) = 1 / (1 + e^-(w*x+b)) logistic regression
# decision boundries z == 0 
## cost for logistic regression : 

## logistic loss function: loss(Y_pred = f(xî, y_real)) = (-y_real * (log(y_pred))) + (1 - y_real)*(-log(1-y_pred)) , y_pred = sigmoid(w*xî+b)
## then cost = 1/m * sumof(loss) her satırın lossu alınır ve toplanır ardından satır sayısına bölünür
## gradient _w and _ b are the same ama y_pred değiştiği için aslında farklı ama yine 1/m sumof(Y-predî.ncsatır - y.inci.satır) * xincisatır j ekelemanı = dwj

```

```python
import numpy as np
import pandas as pd 
import math
```

```python
def sigmoid(w,b,x):
    z = np.dot(x,w) + b 
    sgm = 1/ (1 + np.exp(-z))
    return sgm
```

```python
def current_cost(x,y,w,b):
    y_pred = sigmoid(w,b,x)
    loss = -y*(np.log(y_pred)) - (1-y)*np.log(1-y_pred) 
    cost = np.sum(loss)
    return cost/len(loss)
```

```python
def update_w_b(x,y,m,n,alpha,w_i,b_i):
    y_pred = sigmoid(w_i,b_i,x)
    error = y_pred - y  ## (n,), x in shape (m,n)
    dw =(np.dot(x.T,error)) / m
    db= np.sum(error) / m 
    w_i = w_i - alpha*dw 
    b_i = b_i- alpha*db
    return w_i ,b_i 
    
```

```python
def logistic_regression_model(x,y,m,n,w_i,b_i,alpha,epochs):
    for i in range(epochs):
       w_i,b_i= update_w_b(x,y,m,n,alpha,w_i,b_i)
       if i % 1000 == 0 :
           print(f"iteration: {i}, cost value : {current_cost(x,y,w_i,b_i):.2f}")
    return w_i,b_i
```

```python
def test(x_test, w,b):
    y_pred = sigmoid(w,b,x_test)
    return (y_pred >= 0.5).astype(int)
```

```python
def compute_values(y_pred,y_real):
    result = np.c_[y_pred,y_real]
    print(f"Predictions: , Reals:")
    print(result)

```

```python
def accuracy(y_pred,y_real):
    return np.sum(y_pred==y_real) / len(y_real)
```

```python
data = pd.read_csv('framingham.csv')
y_train = data['TenYearCHD'].values
x_train= data[['male', 'age', 'education', 'currentSmoker', 'cigsPerDay', 'BPMeds',
       'prevalentStroke', 'prevalentHyp', 'diabetes', 'totChol', 'sysBP',
       'diaBP', 'BMI', 'heartRate', 'glucose']]

x_train = x_train.fillna(x_train.mean())

mean = x_train.mean() # (n,m) mean (n,)
std = x_train.std()
x_train_norm = (x_train - mean) / std

x_test = x_train.iloc[17:30,:]
x_test_norm = (x_test - mean) /std

y_test = y_train[17:30]

m,n = x_train.shape
print(m,n)
```

```python
alpha = 0.001
epochs= 10000
w_i = np.zeros(n)
b_i = 0.0 
w_r , b_r = logistic_regression_model(x_train_norm,y_train,m,n,w_i,b_i,alpha,epochs)
predictions = test(x_test_norm,w_r,b_r)
compute_values(predictions,y_test)
print(accuracy(predictions,y_test))
```
