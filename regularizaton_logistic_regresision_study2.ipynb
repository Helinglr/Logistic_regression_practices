```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('train_set.csv')
y_train = data['Ancestry']
data.drop('Ancestry' ,axis= 1, inplace = True)
x_train = data.values
```

```python
def sigmoid(w,x,b):
    z= np.dot(x,w) +b 
    sgm = 1/ (1+ np.exp(-z))
    return sgm
```

```python
def change_set(y_train):
    y_train_changed = [0 if i== 'African' else 1 if i== 'European' else 2 if i == 'EastAsian' else 3 if i== 'Oceanian' else 4 for i in y_train ]
    return y_train_changed
```

```python
def test(w,b,x_test):
    return sigmoid(w,x_test,b)
```

```python
def update(w,b,x,y,lambda_,alpha,m):
    y_pred = sigmoid(w,x,b)
    error= y_pred - y 
    db = np.sum(error) / m 
    dw= np.dot(x.T,error) / m 
    dw_reg = (lambda_/m) * w 
    w = w - alpha*(dw + dw_reg)
    b = b - alpha*db
    return w,b 
```

```python
def logistic_model(x,y,w,b,lambda_,alpha,epochs,m):
    for i in range(epochs):
        w,b= update(w,b,x,y,lambda_,alpha,m)
    return w,b
```

```python

x_test = pd.read_csv('test_set.csv')
y_test = x_test['Ancestry']
x_test.drop('Ancestry', axis=1, inplace= True)


alpha = 0.001 
epochs = 10000
m,n = x_train.shape
w_i = np.zeros(n)
b_i = 0.0
lambda_ = 0.7 
y_train = change_set(y_train)
w,b= logistic_model(x_train,y_train,w_i,b_i,lambda_ , alpha, epochs,m)
```

```python
predictions = test(w,b,x_test)
```

```python
def modify_predictions(pred):
    pred = ['African' if i<=0.2 else 'European' if i<=0.4 else 'EastAsian' if i<=0.6 else 'Oceanian' if i<=0.8 else 'NativeAmerican' for i in pred]
    return pred 
```

```python
predictions = modify_predictions(predictions)
```

```python
def accuracy_score(reals,pred):
    corr = (reals == pred)
    return len(corr)/len(reals) *100
```

```python
print(predictions)
print(accuracy_score(y_test, predictions))
```
