```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('iris.csv')
data.columns
```

```python
def sigmoid(x,w,b): 
    z = np.dot(x,w) + b 
    sgm = 1 / (1 + np.exp(-z))
    return sgm ## sgm (m,)
```

```python
def get_cost(w,b,x,y,m, lambda_):
    ## usualy 
    y_pred = sigmoid(x,w,b)
    loss = -1*y*(np.log(y_pred)) - (1-y)*(np.log(1-y_pred))
    cost = (1/m) * np.sum(loss) 

    ## regularization 
     
    reg = np.sum(w**2)
    reg = reg* lambda_ / (2*m)

    return cost+reg

    
```

```python
def update_w_b_for(w,b,x,y,alpha,lambda_ , m):
    djs= np.zeros(x.shape[1])
    dbs=0.0
    for i in range(m):
        y_pred = sigmoid(x[i],w,b)
        error = y_pred - y[i] ## scalar
        dbs+= error  
        for j in range(x.shape[1]):
            djs[j] = djs[j] + (error*x[i][j])
        
    dbs /= m
    b = b - alpha*dbs 
    for k in range(x.shape[1]):
        djs[k]/=m
   



    dw_reg= np.zeros(x.shape[1])
    for p in range(x.shape[1]):
        dw_reg[p] = ( lambda_/ m ) * w[p]
        w[p]-= (alpha*dw_reg[p]) + (alpha*djs[p])



    return w,b
```

```python
def update_w_b(w,b,x,y,alpha,lambda_ , m):
    ## usualy 
    y_pred = sigmoid(x,w,b)
    error = y_pred - y ## (m,)
    dw= (1/m) * (np.dot(x.T,error)) 
    db= (1/m) * np.sum(error)

    ## reg 
    dw_r = (lambda_/ m) * w

    w = w - alpha*dw_r - alpha*dw
    b = b - alpha*db 
    return w,b
```

```python
def logistic_regression_model(x,y,w,b,m,n,alpha,epochs,lambda_):
    for i in range(epochs):
        w_k,b_k= update_w_b(w,b,x,y,alpha,lambda_,m) ## w vector with 4 features b scalar 
        w_for,b_for = update_w_b_for(w,b,x,y,alpha,lambda_,m)
        w=w_k
        b=b_k
        if i % 1000 == 0 :
            print(f"kiyas of for loop and vector: {w_for ,b_for} -- {w_k,b_k} , ")
            print(f"iter: {i}, cost: {get_cost(w,b,x,y,m , lambda_)}")
    return w,b 
    

```

```python
x_train = data[['x0', 'x1', 'x2', 'x3', 'x4']].iloc[0:102,:]
y_train = data['type']

y_train = y_train.map({'Iris-versicolor': 0 , 'Iris-setosa':1, 'Iris-virginica' : 3 }).values[0:102]
y_test = y_train[38:63]

mean = x_train['x1'].mean()
std = x_train['x1'].std()
x_train['x1_norm'] =(x_train['x1'] - mean) / std
x_train['x1'] = x_train['x1_norm']

x_test = x_train.values[38:63,:]


y_test



```

```python
def test(x_test,w,b):
    y_pred = sigmoid(x_test,w,b)
    return (y_pred >=0.5).astype(int) 
```

```python
alpha = 0.01
epochs = 10000
lambda_ = 0.7
m,n = x_train.shape
w_i = np.zeros(n)
b_i = 0.0
w_f, b_f = logistic_regression_model(x_train.values,y_train,w_i,b_i,m,n,alpha,epochs,lambda_)
```

```python
predictions = test(x_test,w_f,b_f)
predictions = ['Iris-versicolor' if x==0  else 'Iris-setosa' for x in predictions]
y_test = ['Iris-versicolor' if x==0  else 'Iris-setosa' for x in y_test]
result = np.c_[predictions,y_test]
print("Predictions || Reals")
print(result)
```
